```python
def group_age(row):
    if row < 20:
        return 'child'
    elif 20 <= row <= 30:
        return 'young adult'
    elif 30 < row <= 40:
        return 'adult'
    else:
        return 'senior'
```

```python
grouped_age = titanic_train.Age.apply(group_age)
grouped_age = pd.Series(grouped_age, name='GroupedAge')
grouped_age
titanic_train = pd.concat([titanic_train, grouped_age], axis=1)
titanic_train.head()
```

### Finding the best model to go with

```python
# importing ML libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
# Splitting the training data into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y, train_size=.8, random_state=42)

# training the model and making predictions
forest_clf = RandomForestClassifier(random_state=42)
forest_clf.fit(X_train, y_train)
forest_preds = forest_clf.predict(X_val)
forest_roc = roc_auc_score(forest_preds, y_val)
forest_roc
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
knn_preds = knn.predict(X_val)
knn_roc = roc_auc_score(knn_preds, y_val)
knn_roc
```

```python
from xgboost import XGBClassifier

xgb_clf = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_clf.fit(X_train, y_train)
xgb_preds = xgb_clf.predict(X_val)
xgb_roc = roc_auc_score(xgb_preds, y_val)
xgb_roc
```

```python
from sklearn.linear_model import LogisticRegression

lr_clf = LogisticRegression(max_iter=1000)
lr_clf.fit(X_train, y_train)
lr_clf_preds = lr_clf.predict(X_val)
lr_clf_roc = roc_auc_score(lr_clf_preds, y_val)
lr_clf_roc
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train, y_train)
svc_preds = svc.predict(X_val)
svc_roc = roc_auc_score(svc_preds, y_val)
svc_roc
```
